```python
import pandas as pd
```

```python
df=pd.read_csv("customer.csv")
```

```python
df
```

```python
df1=df.iloc[:,2:]
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
x_train, x_test, y_train, y_test = train_test_split(df1.iloc[:,0:2],df.iloc[:,-1],test_size=0.2,random_state=0)
x_train.head()
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
```

```python
x_train.head()
```

# Mentos Jindagi (column Transformer):

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(drop='first', sparse_output=False),['gender']),
    ('tnf2', OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']]),['review', 'education'])
],remainder='passthrough')


#('tnf3', LabelEncoder(), ['purchased'])could have been used but since its output column
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['purchased']),df['purchased'],test_size=0.2)
x_train
```

```python
transformer.fit_transform(x_train)
```
